Because of this, we will filter out those pairs that have less than 150 flights.

```python
# Keep only the rows where the pair has over 150 flights
filtered_flights = flights.groupby( ['Departure City', 'Arrival City']).filter( lambda x: len(x) >= 150 )
print("There are " + str(filtered_flights.shape[0]) + " rows.")
print("There are " + str(len(filtered_flights.groupby( ['Departure City', 'Arrival City']).size())) + " pairs of cities.")
```

To further, we also want to generally remove any outliers from our numerical columns. To do so, we can apply the Interquantile Range (IQR) Method where we define outliers as values that are below (Q1 - 1.5 x IQR) and above (Q3 + 1.5 x IQR). Here Q1 is defined as the 25th percentile and Q3 is defined as the 75th percentile while IQR is the range within the two. Below we will define a function that helps us handle outliers within our data.

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Return only the rows where the value in the column is within the bounds
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

Now, lets use this function to filter out all of the outliers of numerical columns in our data. Additionally, doing some research shows us that private jets can hold up to 19 people in the biggest jet but since your typical traveler will not be flying on private yets, lets exclude all flights where we have less than 20 passengers in case they are private jets.

```python
# Apply the remove_outliers function to the relevant columns
filtered_flights = remove_outliers(filtered_flights, 'Fare')
filtered_flights = remove_outliers(filtered_flights, 'Passengers')
filtered_flights = remove_outliers(filtered_flights, 'Miles')


# Remove rows where there are less than 20 passengers
filtered_flights = filtered_flights[filtered_flights['Passengers'] >= 20]
```

With this, we see that we have cut our dataset down by nearly 100,000 rows and gotten rid of the highs and lows. While this is certainly a lot to cut, this is actually not much of a problem since we still have 146,000 rows to work with to make our model which should be plenty of data.

## Creating a Model

Now that we've thoroughly explored what our dataset has to offer, we can create a machine learning model using scikitlearn's library to predict airplane ticket fares. This includes splitting data, preprocessing, training, hyperparameter tuning, and finally, creating and evaluating our final model.

```python
# Imports
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import time
import joblib
import warnings
```

### Splitting up the data

Time to split up the data! We will need a training set to train the model on, a validation set for hyperparameter tuning, and lastly, a test set to evaluate final model performance. We will take a 70%/15%/15% split for our sets. Our features we want to train on are year, quarter, miles, passengers, popularity, departure city, and arrival city. Our target is the fare.

```python
# Define our splits
train_split = 0.7
val_split = 0.15
test_split = 0.15

# Features
X_features = filtered_flights[['Year', 'Quarter', 'Miles', 'Passengers', 'Popularity', 'Departure City', 'Arrival City']]
Y_feature = filtered_flights['Fare']

# Split into a test set and a broader set ( training + validation ), setting seed for reproducability
x_remaining, x_test, y_remaining, y_test = train_test_split( X_features, Y_feature, test_size=test_split, random_state=27)

# Calculate adjustments for next split
remaining_amt = 1 - test_split
val_split_adjusted = val_split / remaining_amt

# Get train and validation split
x_train, x_val, y_train, y_val = train_test_split( x_remaining, y_remaining, test_size=val_split_adjusted, random_state=27)

# Check our results
print( "x_train: " + str( x_train.shape ) + "\n"
       "x_val: " + str( x_val.shape ) + "\n"
       "x_test: " + str( x_test.shape ) + "\n"
       "y_train: " + str( y_train.shape ) + "\n"
       "y_val: " + str( y_val.shape ) + "\n"
       "y_test: " + str( y_test.shape ) )
```

### Preprocessing

Next, we need to preprocess our data. We have nominal categorical variables in the form of "Quarter", "Departure City", and "Arrival City". We want to turn these into one hot vectors so that our models can process them, and avoid any potential issues of passing in the unordered variable directly. We also have various kinds of numeric data. For instance, "Miles" can have magnitude of 1000, whereas popularity has a magnitude of 10. Normalizing these with z scores will allow us to make our model more accurate.

```python
# Split into categorical and numerical features
cat_features = [ 'Quarter', 'Departure City', 'Arrival City']
num_features = [ 'Year', 'Miles', 'Passengers', 'Popularity']

# Initialize preprocessors
cat_preproc = OneHotEncoder()
num_preproc = StandardScaler()

# Consoldiate into one preprocessor
preprocessor = ColumnTransformer( [ ('categorical', cat_preproc, cat_features), 
                                    ('numerical', num_preproc, num_features)])
```

### Training and Pipelines

Now that we've prepared everything, we can train our models! We'll need to create pipelines which will simplify our workflow and combines our preprocessor and our regressor models. We've chosen to use 3 models in particular. A linear regression will be our benchmark for our other two models. This is because we suspect that our data is nonlinear, which means that a linear regression won't fit too well. We can also expect that if we chose a wrong model, or if it doesn't perform well enough, it will likely be closer to the linear model. For our nonlinear models, we chose a random forest regressor and a k-nearest neighbors. A random forest regressor is robust to handle complex datasets (nonlinear), and predicts continous outputs like our fares. K-nearest neighbors will work locally, meaning that it will look for similarly grouped data and use that as a means to identify it. We think that these models will best perform on our dataset.

Also we will use Root Mean Squared Error (RMSE) and Mean Absolute Error (MAE) as our metrics due to the nonlinearity of our dataset. We will also measure $R^2$ because of our suspicions of nonlinearity. $R^2$ is inherently linear due to coming from Pearson's correlation coefficient (r), so similar to using linear regression, it may give us a good comparison for what a poorer performance looks like.

```python
# Create DataFrame to store our results in
results = pd.DataFrame(columns=['Model', 'Mean Absolute Error', 'Root Mean Squared Error', 'R2', 'Time Elapsed (Minutes)'])
results

# Useful to store our models for later
pipelines = {}
# Define our set of regressors
models = {
           'LinearRegression': LinearRegression(),
           'RandomForestRegressor': RandomForestRegressor( n_estimators=50, n_jobs=-1, random_state=27),
           'KNeighborsRegressor': KNeighborsRegressor(n_neighbors=5) 
         }



# Loop through each regressor
for model_name, model in models.items():
    pipeline = Pipeline( steps=[ ('preprocessing', preprocessor),
                                 ( 'regression', model)]) 
                                                           
    start_time = time.time()

    pipeline.fit( x_train, y_train) # Train model

    pipelines[ model_name ] = pipeline # Save model for later

    yhat_val = pipeline.predict( x_val ) # predict fares on validation set

    end_time = time.time()

    # Calculate metrics for accuracy

    val_R2 = round( r2_score( y_val, yhat_val), 6 )
    val_MAE = mean_absolute_error( y_val, yhat_val)
    val_RMSE = np.sqrt(mean_squared_error( y_val, yhat_val))
    
    elapsed_time = (end_time - start_time) / 60  # Added elapsed time for the sake of knowing how long to expect to wait

    # Create a dictionary with model results
    row = {
            'Model': model_name,
            'Mean Absolute Error': val_MAE,
            'Root Mean Squared Error': val_RMSE,
            'R2': val_R2,
            'Time Elapsed (Minutes)': elapsed_time
          }

    # Add results to dataframe
    results = pd.concat( [results, pd.DataFrame([row])], ignore_index=True )
    
    
results = results.set_index('Model')
results
```

Let's try to interpret this. The mean absolute error is the difference between the actual and predicted value. In other words, we can directly interpret this metric as the average dollars our model was off from the actual fare price. Our linear regression, random forest, and K Neighbors performed with an average \$37, \\$18, \\\$25.

Our RMSE isn't interpreted as easily. Although it can be referred to in a similar manner as MAE, that is, “RMSE reflects the average amount the predicted fare was off from the actual,” due to the way it’s calculated it penalizes outliers and higher errors more. Because of this, we can always expect our RMSE to be greater than or equal to our MAE. As a comparison with MAE, the greater the difference is between them, the more we can see the effect of bigger outliers in our data.

Lastly, the $R^2$ will tell us how much the variance our model can explain by our measured variables. Since 1.0 is a perfectly captured relationship, something like random forest’s 0.82 captures the relationship well and the variables are being modeled efficiently. However, the linear regression with 0.48 isn’t capturing the relationship as well, and suggests that we may be better of utilizing the other models.

### Hyperparameter tuning

As expected, our linear regressor appears to be performing the worst among our 3 models, while our random forest is doing the best. This is is only with one set of hyperparameters though. Let's do some fine tuning to see if we can improve our performance! Linear regression models aren't highly tunable, so we'll be finding the best hyperparameters for our RandomForestRegressor and KneighborsRegressor. Tuning both of these takes around 8 hours, so we went ahead and saved the models to a file for easy retrival.

```python
# Useful to store our models for later
tuned_models = {}

# Define our set of regressors & potential parameters
models = { 
           'RandomForestRegressor': RandomForestRegressor( n_jobs=-1, random_state=27),
           'KNeighborsRegressor': KNeighborsRegressor(n_neighbors=5) 
         }
model_params = {
    'RandomForestRegressor': {
        'regression__n_estimators': [50, 75, 100],
        'regression__max_depth': [5, 10, None ],
        'regression__min_samples_split': [2, 5, 10]
    },
    'KNeighborsRegressor': {
        'regression__n_neighbors': [3, 5, 10],
        'regression__weights': ['uniform', 'distance'],
        'regression__p': [1,2]
    }
}

# Loop through each regressor
for model_name, model in models.items():
    pipeline = Pipeline( steps=[ ('preprocessing', preprocessor),
                                 ( 'regression', model)]) 

    # Retrive current model's parameter dict.
    model_param = model_params[ model_name ]
    
    random_search = RandomizedSearchCV(
        pipeline,
        model_param,
        n_iter=10, # make 10 iterations of each regressor
        cv=5,
        scoring='neg_mean_absolute_error',
        n_jobs=-1,
        random_state=27,
        verbose=1
    )

    # Train & predict (one step)
    random_search.fit( x_train, y_train )

    # Save models to a dictionary
    tuned_models[ model_name ] = random_search

# Save our models to a file outside our notebook, so we can avoid running this for 8 hours every time
joblib.dump( tuned_models, 'data/tuned_models.pkl')
```

```python
tuned_models = joblib.load( 'data/tuned_models.pkl')
print('success!')
```

Looks like we were able to get it to train and save successfully! Let's check the results of that hard work.

```python
# Create new DataFrame to store our best results in
best_results = pd.DataFrame(columns=['Model', 'Best Parameters', 'Cross Validation Score'])

models = [ 'RandomForestRegressor', 'KNeighborsRegressor' ]

# Add results of models to a df
for model_name in models:
    model = tuned_models[model_name]
    
    row = {
        'Model': model_name,
        'Best Parameters': model.best_params_,
        'Cross Validation Score': -model.best_score_ # undo negative score
    }

    best_results = pd.concat( [best_results, pd.DataFrame([row])], ignore_index=True )
    
# Cleaning up appearances
pd.set_option('display.max_colwidth', None)
warnings.filterwarnings('ignore', message='.*behavior of DataFrame concatenation.*')

# Print
best_results
```

Best parameters? Best cross validation score? How can we interpret those?

To explain these, we need to understand how the RandomizedSearchCV works. The randomized search cross validation (RSCV) is given a set of possible hyperparameters to tune different iterations of models on. Where it differs from its sister method GridSearchCV, is that instead of making an exhuastive list of models for every possible combination, it randomly picks the hyperparameters and makes n_iter models, which we set as 10. The reason for doing it this way rather than GridSearchCV was a choice of computation speed over model performance. Training the 10 iterations for each model alone took around 8 hours, increasing that by nearly 3 times would increase the training time to nearly 24 hours! To keep a smooth workflow and reduce the risk of possible errors, we decided that the shorter training time at the potential cost of model performance was worth it. 

Best parameters shows the hyperparameters for the model that performed best on the cross validation score. If we take a look at it, it makes sense. For example, our Random Forest Regressor looks like it had a very detail oriented model. n_estimators is the number of trees in the model. More trees equals a more in depth analysis, so of our possible options, a max of 100 should have led to the best results. Likewise for the max_depth. By not capping how deep the trees can go, we let it make a complex modeling of our feeatures.

RSCV creates 5 splits in the given data and creates 5 different models and evaluates their performance on one of the splits. It then evaluates its performance across all 5 and averages it (using an MAE metric), into what we see as our cross validation score. In simple terms, For our best model -- Random Forest Regressor -- it had an average of $27 error on our cross validations.

Let's check out one final performance of these regressors on our validation set.

```python
results = pd.DataFrame(columns=['Model', 'Mean Absolute Error', 'Root Mean Squared Error', 'R2'])

for model_name in tuned_models:

    # Grab model
    model = tuned_models[model_name]

    # Evaluate on validation set
    yhat_val = model.predict( x_val )
    mae = mean_absolute_error( y_val, yhat_val )
    rmse = np.sqrt( mean_squared_error( y_val, yhat_val ) )
    r2 = r2_score( y_val, yhat_val )
    
    # Create a dictionary with model results
    row = {
            'Model': model_name,
            'Mean Absolute Error': mae,
            'Root Mean Squared Error': rmse,
            'R2': r2,
          }

    results = pd.concat([results, pd.DataFrame([row])], ignore_index=True)

results.set_index('Model', inplace=True)
results
```

Not too bad! It appears that in every metric our random forest regressor is performing the best. We've reduced our average error (MAE) from \$18 to \\$12 through tuning. Likewise RMSE has decreased from 27.2 to 19.2. This suggests that we were able to find a good set of hyperparameters to work with for our final model.
